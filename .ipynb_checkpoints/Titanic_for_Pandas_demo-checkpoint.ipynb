#  Apply a function to a column to generate new data

```python
def checkTitles(name):
    if "Mr." in name:
        return 0
    elif "Mrs." in name:
        return 1
    elif "Miss." in name:
        return 2
    elif "Master." in name:
        return 3
    else:
        return 4
```

```python
df["Title"] = df.Name.apply(checkTitles)
```

```python
df.head()
```

```python

sns.set(style="darkgrid")

ax = sns.countplot(x="Title", data=df[df.Survived>-1], hue = "Survived")
```

## take the initial letter of Cabin

```python
def Cabin_conversion(cabin):
    if str(cabin)=='nan':
        return 0
    else:
        return cabin[0]
```

# Predict missing ages

```python
from sklearn.cross_validation import train_test_split
from sklearn import metrics
```

```python
age_train = df_dummy.dropna()
```

```python
age_test = df_dummy[np.isnan(df_dummy.Age)]
```

```python
age_train.head()
```

```python
age_test.head()
```

```python
y_train = age_train.pop("Age")
```

```python
x_train = age_train.values
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(x_train, y_train, random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
params = {'n_estimators': 200, 'max_depth': 17, 'min_samples_split': 1,
          'learning_rate': 0.01, 'loss': 'ls'}
clf = GradientBoostingRegressor(**params)
clf.fit(Xtrain, ytrain)
ypred2 = clf.predict(Xtest)
```

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())


```

```python
print(rmse(ytest, ypred2))
plt.scatter((ytest), (ypred2))
plt.xlabel('prediction')
plt.ylabel('actual')


```

```python
from sklearn import linear_model
clf = linear_model.Lasso(alpha = 0.1)
clf.fit(Xtrain, ytrain)
prediction = clf.predict(Xtest)
print(rmse(ytest, prediction))
plt.scatter(prediction, ytest)
plt.xlabel('prediction')
plt.ylabel('actual')
```

```python
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor(n_estimators=100, random_state=0)
clf.fit(Xtrain, ytrain)
ypred1 = clf.predict(Xtest)
```

```python
print(rmse(ytest, ypred1))
plt.scatter(ypred1, ytest)
plt.xlabel('prediction')
plt.ylabel('actual')
```

```python
from sklearn.ensemble import GradientBoostingRegressor
params = {'n_estimators': 300, 'max_depth': 6, 'min_samples_split': 1,
          'learning_rate': 0.01, 'loss': 'ls'}
clf2 = GradientBoostingRegressor(**params)
clf2.fit(Xtrain, ytrain)
ypred2 = clf2.predict(Xtest)
```

```python
print(rmse(ytest, ypred2))
plt.scatter((ytest), (ypred2))
plt.xlabel('prediction')
plt.ylabel('actual')
```
